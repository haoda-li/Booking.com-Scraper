```python
def process_info(table, fw):
    for row in table:
        if len(row) >= 5:
            roomtype = row[0].split("\n")[0].replace("\n", "")
            price = row[2].split("\n")[0].replace("\n","")
            left = row[4].split("\n")[-4].replace("\n", "")
            fw.write(roomtype+"|"+price+"|"+left+"\n")
```
